```python
"""
- load iris dataset
- normalize dataset
- compute covariance matrix
- compute eigenvals / eigenvecs
"""
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
data = load_iris()['data']
targets = load_iris()['target']



class PCA:
    
    def __init__(self,K=2):
        self.K = K
        
    def _sorted_eig(self,mat):
        eigen_vals, eigen_vecs = np.linalg.eig(mat)
        eigen_vecs = eigen_vecs.T
        eigen_vals,eigen_vecs = (np.array(t) for t in zip(*sorted(zip(eigen_vecs,eigen_vecs),reverse=True)))
        return eigen_vals,eigen_vecs

    def fit_transform(self,data):
        n = data.shape[0]
        # normalize
        X = (data-data.mean(0))/(data.std(0)+1e-6)
        # compute covariance 
        cov = X.T @ X / n
        # compute eigenvalues and eigenvectors
        eigen_vals,eigen_vecs = self._sorted_eig(cov)
        # select K highest components
        eigen_vals, eigen_vecs = eigen_vals[:K], eigen_vecs[:K]
        # project data onto eigenspace
        X = X @ eigen_vecs.T
        return X
        
model = PCA(2)
x = model.fit_transform(data)
plt.scatter(x[:,0],x[:,1],c=targets)
```

```python


class PCA:
    
    def __init__(self,n_components=2):
        self.n_components = n_components
        

    def fit_transform(self,data):
        K = self.n_components
        N = data.shape[0]
        # normalize data
        data = (data-data.mean(0))/(data.std(0) +1e-6)    
        # get covariance matrix
        Sigma = np.cov(data.T)
       
        # get eigen vals and vectors of covariance

        # be careful here, evecs are returned in columns not in rows
        # this can mess things up if you pick of vecs by row vs col
        evals, evecs = np.linalg.eig(Sigma)
        
        principal_idxs = np.flip(evals.argsort())[:self.n_components]
        principal_evecs = evecs[:,principal_idxs]
        
        # NOTE: this may output a mirror image of sklearn's method
        # that's ok, since eigenvectors are scaled to have unit length
        # eigenvals can be -/+ so mirror just means we picked a different sign
        # than sklearn
        return (data @ principal_evecs)

    
iris_data = load_iris()
a,b = iris_data['data'],iris_data['target']
pca = PCA(2)
proj = pca.fit_transform(a)
plt.scatter(proj[:,0],proj[:,1],c=b)
```
